```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


def load_data(path):
    """Load Yelp business JSON (JSON Lines format)."""
    return pd.read_json(path, lines=True)


def filter_restaurants(df):
    """Keep only rows where categories contain 'Restaurants'."""
    df = df.copy()
    df["categories"] = df["categories"].fillna("")
    return df[df["categories"].str.contains("Restaurants", case=False, na=False)].copy()


def flatten_column(df, column, prefix):
    """Flatten nested dict columns like attributes or hours."""
    if column not in df.columns:
        return df
    expanded = pd.json_normalize(df[column].fillna({}))
    expanded.columns = [f"{prefix}{col}" for col in expanded.columns]
    df = df.drop(columns=[column]).reset_index(drop=True)
    return pd.concat([df, expanded.reset_index(drop=True)], axis=1)


def clean_types(df):
    """Fix numeric and boolean columns."""
    df = df.copy()

    for col in ["stars", "review_count", "latitude", "longitude"]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    if "is_open" in df.columns:
        df["is_open"] = pd.to_numeric(df["is_open"], errors="coerce").map({0: False, 1: True})

    return df


def basic_cleanup(df):
    """Remove duplicates and normalize blanks."""
    df = df.copy()

    if "business_id" in df.columns:
        df = df.drop_duplicates(subset=["business_id"])

    for col in ["name", "address", "city", "state", "postal_code"]:
        if col in df.columns:
            df[col] = df[col].replace("", np.nan)

    return df


def quick_visuals(df):
    """Simple sanity-check plots."""
    if "stars" in df.columns:
        plt.figure()
        df["stars"].dropna().plot(kind="hist", bins=20)
        plt.title("Restaurant Star Distribution")
        plt.xlabel("Stars")
        plt.show()

    if "review_count" in df.columns:
        plt.figure()
        np.log1p(df["review_count"].dropna()).plot(kind="hist", bins=30)
        plt.title("log(1 + Review Count)")
        plt.xlabel("log1p(review_count)")
        plt.show()
```

```python
def flatten_dict_column(df, col, prefix):
    if col not in df.columns:
        return df

    series = df[col].apply(lambda x: {} if pd.isna(x) else x)
    expanded = pd.json_normalize(series)
    expanded.columns = [f"{prefix}{c}" for c in expanded.columns]
    df = df.drop(columns=[col]).reset_index(drop=True)
    df = pd.concat([df, expanded], axis=1)

    return df
```

```python
BUSINESS_JSON = "data/00-raw/raw_yelp_academic_dataset_business.json"
OUT_CSV = "data/02-processed/restaurants_clean.csv"

# Load
df = load_data(BUSINESS_JSON)

# Filter
restaurants = filter_restaurants(df)

# Flatten nested columns
restaurants = flatten_column(restaurants, "attributes", "attr_")
restaurants = flatten_column(restaurants, "hours", "hours_")

# Clean types
restaurants = clean_types(restaurants)

# Basic cleanup
restaurants = basic_cleanup(restaurants)

print("Total rows:", len(df))
print("Restaurant rows:", len(restaurants))
print(restaurants[["stars", "review_count"]].describe())

# Visualize
quick_visuals(restaurants)

# Save
restaurants.to_csv(OUT_CSV, index=False)
print("Saved to:", OUT_CSV)
```
