## Research Question

__What attributes in San Diegan restaurants are associated with higher reviews on Yelp?__

## Background and Prior Work

San Diego is a growing hub for food, with recent data estimating approximately 5,000 restaurants across the county.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Highly regarded for its quality Mexican cuisine, San Diego also hosts a variety of different restaurants with popular spots for Asian food, seafood, fusion, fast food, coffee, and desserts. But amidst the vast array of options available, what sets apart a local favorite from a hard “maybe”? Is it outdoor seating to take in the unique climate? Takeout options to expedite food pickups during rush hour? Or pure ambiance to draw in curious passerbys? 

In this project we aim to analyze Yelp data from restaurants in San Diego to identify key features that contribute to a restaurant’s success. We hope our results will provide a better understanding of the San Diego restaurant-goer demographic, from tourists to locals, so that businesses can understand the best way to cater to and enhance the dining experience for their patrons.

To begin our search for background data we explored a past project with a broader scope of restaurants in the US. It identified differences in rating distributions based on cuisine type, with statistical tests rejecting the possibility of these differences occurring by chance.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) In the context of our project, this motivated us to pursue cuisines as a possible attribute for a restaurant’s success here in San Diego. Analyzing associations between cuisine type and rating may give insight into which types of cuisines tend to perform well. Looking at San Diego's population census with 29.6% Hispanic and 17.2% Asian being the highest non-white ethnicities, we predict these cuisines will be associated with high ratings.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) San Diego has notable Vietnamese and Filipino populations, so we will focus on these types of cuisine.

Restaurant attributes such as service quality, price, and atmosphere are broadly applicable across restaurant types and are particularly relevant to online review platforms such as Yelp. Prior research, including Chun (2020), has examined how key restaurant attributes such as food quality, service quality, price, and atmosphere influence customer satisfaction in Mongolia. Using survey-based data and statistical analysis, Chun (2020) found that food quality and service quality exhibit the strongest positive correlations with customer satisfaction.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

There has been prior research on what attributes on Yelp can contribute to a restaurant’s success. Studies using data from Yelp have found that data is not random and there are common denominators for success. One example found that across the US, restaurants have significant differences in cuisines which show that certain cuisines perform dramatically better than others because of customer’s preferences. Additionally, factors such as delivery options, reservations, and group accommodations were associated with higher star ratings on Yelp. These observations could potentially be found in the San Diego market as well. 

Food quality was identified as the strongest predictor of satisfaction, even when controlling for other factors such as price and atmosphere. In the context of this project, food quality can be approximated using information extracted from Yelp review text, including sentiment, tone, and the presence of positive or negative words and phrases. Service quality, defined in the study as staff friendliness and speed of service, also demonstrated a strong positive relationship with customer satisfaction. On Yelp, service quality may be indirectly measured through business attributes such as Good for Groups, Reservations, and Delivery.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)

The weakest, though still positive, correlation observed in the study was associated with the atmosphere and the physical dining environment. Chun (2020) suggests this may be due to variation across restaurant types, such as dine-in versus takeout establishments. Yelp attributes related to atmosphere, including Outdoor Seating, Noise Level, Has TV, and Ambience, can therefore be analyzed in this project to assess whether similar patterns are observed among restaurants in San Diego.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


def load_data(path):
    """Load Yelp business JSON (JSON Lines format)."""
    return pd.read_json(path, lines=True)


def filter_restaurants(df):
    """Keep only rows where categories contain 'Restaurants'."""
    df = df.copy()
    df["categories"] = df["categories"].fillna("")
    return df[df["categories"].str.contains("Restaurants", case=False, na=False)].copy()


def flatten_column(df, column, prefix):
    """Flatten nested dict columns like attributes or hours."""
    if column not in df.columns:
        return df
    expanded = pd.json_normalize(df[column].fillna({}))
    expanded.columns = [f"{prefix}{col}" for col in expanded.columns]
    df = df.drop(columns=[column]).reset_index(drop=True)
    return pd.concat([df, expanded.reset_index(drop=True)], axis=1)


def clean_types(df):
    """Fix numeric and boolean columns."""
    df = df.copy()

    for col in ["stars", "review_count", "latitude", "longitude"]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    if "is_open" in df.columns:
        df["is_open"] = pd.to_numeric(df["is_open"], errors="coerce").map({0: False, 1: True})

    return df


def basic_cleanup(df):
    """Remove duplicates and normalize blanks."""
    df = df.copy()

    if "business_id" in df.columns:
        df = df.drop_duplicates(subset=["business_id"])

    for col in ["name", "address", "city", "state", "postal_code"]:
        if col in df.columns:
            df[col] = df[col].replace("", np.nan)

    return df


def quick_visuals(df):
    """Simple sanity-check plots."""
    if "stars" in df.columns:
        plt.figure()
        df["stars"].dropna().plot(kind="hist", bins=20)
        plt.title("Restaurant Star Distribution")
        plt.xlabel("Stars")
        plt.show()

    if "review_count" in df.columns:
        plt.figure()
        np.log1p(df["review_count"].dropna()).plot(kind="hist", bins=30)
        plt.title("log(1 + Review Count)")
        plt.xlabel("log1p(review_count)")
        plt.show()
```

```python
def flatten_dict_column(df, col, prefix):
    if col not in df.columns:
        return df

    series = df[col].apply(lambda x: {} if pd.isna(x) else x)
    expanded = pd.json_normalize(series)
    expanded.columns = [f"{prefix}{c}" for c in expanded.columns]
    df = df.drop(columns=[col]).reset_index(drop=True)
    df = pd.concat([df, expanded], axis=1)

    return df
```

```python
BUSINESS_JSON = "data/00-raw/raw_yelp_academic_dataset_business.json"
OUT_CSV = "data/02-processed/restaurants_clean.csv"

# Load
df = load_data(BUSINESS_JSON)

# Filter
restaurants = filter_restaurants(df)

# Flatten nested columns
restaurants = flatten_column(restaurants, "attributes", "attr_")
restaurants = flatten_column(restaurants, "hours", "hours_")

# Clean types
restaurants = clean_types(restaurants)

# Basic cleanup
restaurants = basic_cleanup(restaurants)

print("Total rows:", len(df))
print("Restaurant rows:", len(restaurants))
print(restaurants[["stars", "review_count"]].describe())

# Visualize
quick_visuals(restaurants)

# Save
restaurants.to_csv(OUT_CSV, index=False)
print("Saved to:", OUT_CSV)
```
