# Research Question

__What attributes in Santa Barbara restaurants are associated with higher reviews on Yelp?__

## Background and Prior Work

Santa Barbara is a growing hub for food, with recent data estimating approximately 5,000 restaurants across the county.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Highly regarded for its quality Mexican cuisine, Santa Barbara also hosts a variety of different restaurants with popular spots for Asian food, seafood, fusion, fast food, coffee, and desserts. But amidst the vast array of options available, what sets apart a local favorite from a hard “maybe”? Is it outdoor seating to take in the unique climate? Takeout options to expedite food pickups during rush hour? Or pure ambiance to draw in curious passerbys? 

In this project we aim to analyze Yelp data from restaurants in Santa Barbara to identify key features that contribute to a restaurant’s success. We hope our results will provide a better understanding of the Santa Barbara restaurant-goer demographic, from tourists to locals, so that businesses can understand the best way to cater to and enhance the dining experience for their patrons.

To begin our search for background data we explored a past project with a broader scope of restaurants in the US. It identified differences in rating distributions based on cuisine type, with statistical tests rejecting the possibility of these differences occurring by chance.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) In the context of our project, this motivated us to pursue cuisines as a possible attribute for a restaurant’s success here in Santa Barbara. Analyzing associations between cuisine type and rating may give insight into which types of cuisines tend to perform well. Looking at Santa Barbara's population census with 63% White being the highest ethnicity, we predict these cuisines will be associated with high ratings.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Santa Barbara has notable Hispanic populations, so we will focus on this type of cuisine as well.

Restaurant attributes such as service quality, price, and atmosphere are broadly applicable across restaurant types and are particularly relevant to online review platforms such as Yelp. Prior research, including Chun (2020), has examined how key restaurant attributes such as food quality, service quality, price, and atmosphere influence customer satisfaction in Mongolia. Using survey-based data and statistical analysis, Chun (2020) found that food quality and service quality exhibit the strongest positive correlations with customer satisfaction.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

There has been prior research on what attributes on Yelp can contribute to a restaurant’s success. Studies using data from Yelp have found that data is not random and there are common denominators for success. One example found that across the US, restaurants have significant differences in cuisines which show that certain cuisines perform dramatically better than others because of customer’s preferences. Additionally, factors such as delivery options, reservations, and group accommodations were associated with higher star ratings on Yelp. These observations could potentially be found in the Santa Barbara market as well. 

Food quality was identified as the strongest predictor of satisfaction, even when controlling for other factors such as price and atmosphere. In the context of this project, food quality can be approximated using information extracted from Yelp review text, including sentiment, tone, and the presence of positive or negative words and phrases. Service quality, defined in the study as staff friendliness and speed of service, also demonstrated a strong positive relationship with customer satisfaction. On Yelp, service quality may be indirectly measured through business attributes such as Good for Groups, Reservations, and Delivery.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)

The weakest, though still positive, correlation observed in the study was associated with the atmosphere and the physical dining environment. Chun (2020) suggests this may be due to variation across restaurant types, such as dine-in versus takeout establishments. Yelp attributes related to atmosphere, including Outdoor Seating, Noise Level, Has TV, and Ambience, can therefore be analyzed in this project to assess whether similar patterns are observed among restaurants in Santa Barbara.
