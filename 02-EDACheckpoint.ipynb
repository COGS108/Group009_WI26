## Background and Prior Work

The United States has one of the largest and most diverse restaurant industries in the world, with hundreds of thousands of establishments offering a wide range of cuisines, service styles, and dining experiences. With so many options available to consumers, online review platforms such as Yelp play a significant role in shaping customer decisions. Yelp allows users to rate restaurants and provide feedback based on their dining experience, creating large datasets that can be analyzed to better understand what factors contribute to a restaurant’s success. Attributes such as service quality, price range, ambience, and available amenities may influence how customers perceive restaurants and how they rate them online.

In this project, we analyze Yelp data from restaurants across the United States to identify key features that contribute to higher ratings and customer satisfaction. By examining business attributes such as delivery options, reservations, outdoor seating, WiFi access, ambience, and accessibility features, we aim to determine which operational and environmental characteristics are most strongly associated with higher Yelp ratings. Understanding these relationships can help restaurant owners and managers better tailor their services and amenities to meet customer expectations.

To begin our search for background data we explored a past project with a broader scope of restaurants in the US. It identified differences in rating distributions based on cuisine type, with statistical tests rejecting the possibility of these differences occurring by chance.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Prior research has explored how different restaurant attributes influence customer satisfaction and online ratings. Studies using Yelp data have shown that ratings are not randomly distributed and that certain restaurant characteristics tend to be associated with stronger performance. 

Restaurant attributes such as service quality, price, and dining atmosphere are widely recognized as important contributors to customer satisfaction. Chun (2020) examined how restaurant attributes influence satisfaction, revisit intention, and customer recommendations, finding that food quality and service quality demonstrate strong positive correlations with customer satisfaction. While food quality itself is difficult to measure directly in large datasets, proxy indicators such as amenities, service options, and environment-related attributes may capture aspects of the overall dining experience that influence customer ratings.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

There has been prior research on what attributes on Yelp can contribute to a restaurant’s success. Studies using data from Yelp have found that data is not random and there are common denominators for success. One example found that across the US, restaurants have significant differences in cuisines which show that certain cuisines perform dramatically better than others because of customer’s preferences. Additionally, factors such as delivery options, reservations, and group accommodations were associated with higher star ratings on Yelp. These observations could potentially be found in the US market as well. 

Food quality was identified as the strongest predictor of satisfaction, even when controlling for other factors such as price and atmosphere. In the context of this project, food quality can be approximated using information extracted from Yelp review text, including sentiment, tone, and the presence of positive or negative words and phrases. Service quality, defined in the study as staff friendliness and speed of service, also demonstrated a strong positive relationship with customer satisfaction. On Yelp, service quality may be indirectly measured through business attributes such as Good for Groups, Reservations, and Delivery.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)

Additionally, other studies analyzing Yelp data have identified operational features such as delivery availability, reservation options, and group accommodations as factors that can influence restaurant performance. These attributes reflect convenience and service accessibility, which are important elements of the modern dining experience. Environmental attributes, including ambience and outdoor seating, may also shape customer perceptions of the restaurant atmosphere and contribute to higher ratings.

By analyzing these attributes across a large sample of restaurants in the United States, this project aims to identify patterns between restaurant characteristics and Yelp ratings. The results may provide insights into which amenities and operational features are most strongly associated with positive customer evaluations in the U.S. restaurant market.

### Dataset #1: Yelp Open Dataset

Here we take the Yelp API data from the Yelp Open Dataset, filter it to only restaurants, then clean the data.

```python
import json
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
import scipy.stats as stats
from scipy.stats import ttest_ind, chisquare, normaltest
```

```python
def load_data(path):
    rows = []
    bad = 0

    with open(path, "r", encoding="utf-8", errors="replace") as f:
        for line_num, line in enumerate(f, start=1):
            line = line.strip()
            if not line:
                continue
            try:
                obj = json.loads(line)
                if isinstance(obj, dict) and "business_id" in obj:
                    rows.append(obj)
            except json.JSONDecodeError:
                bad += 1
                if bad <= 3:
                    print(f"Bad JSON on line {line_num}: {line[:200]}")

    print("Loaded records:", len(rows), "| Bad lines skipped:", bad)
    return pd.DataFrame.from_records(rows)


def filter_restaurants(df):
    df = df.copy()
    df["categories"] = df["categories"].fillna("")
    return df[df["categories"].str.contains("Restaurants", case=False, na=False)].copy()


def flatten_column(df, column, prefix):
    if column not in df.columns:
        return df
    expanded = pd.json_normalize(df[column].fillna({}))
    expanded.columns = [f"{prefix}{col}" for col in expanded.columns]
    df = df.drop(columns=[column]).reset_index(drop=True)
    return pd.concat([df, expanded.reset_index(drop=True)], axis=1)


def clean_types(df):
    df = df.copy()

    for col in ["stars", "review_count", "latitude", "longitude"]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    if "is_open" in df.columns:
        df["is_open"] = pd.to_numeric(df["is_open"], errors="coerce").map({0: False, 1: True})

    return df


def basic_cleanup(df):
    df = df.copy()

    if "business_id" in df.columns:
        df = df.drop_duplicates(subset=["business_id"])

    for col in ["name", "address", "city", "state", "postal_code"]:
        if col in df.columns:
            df[col] = df[col].replace("", np.nan)

    return df
```

```python
def flatten_dict_column(df, col, prefix):
    if col not in df.columns:
        return df

    series = df[col].apply(lambda x: {} if pd.isna(x) else x)
    expanded = pd.json_normalize(series)
    expanded.columns = [f"{prefix}{c}" for c in expanded.columns]
    df = df.drop(columns=[col]).reset_index(drop=True)
    df = pd.concat([df, expanded], axis=1)

    return df
```

```python
BUSINESS_JSON = "data/00-raw/raw_yelp_academic_dataset_business.json"
OUT_CSV = "data/02-processed/restaurants_clean.csv"

df = load_data(BUSINESS_JSON)

restaurants = filter_restaurants(df)

restaurants = flatten_column(restaurants, "attributes", "attr_")
restaurants = flatten_column(restaurants, "hours", "hours_")

restaurants = clean_types(restaurants)

restaurants = basic_cleanup(restaurants)

restaurants.to_csv(OUT_CSV, index=False)
print("Saved to:", OUT_CSV)
```

## Results

### Exploratory Data Analysis


#### Stars and Review Counts

We're going to start by looking at our data distributions for star ratings, then seeing what role review counts play in predicting a given restaurant's ratings.

```python
plt.figure(figsize=(8, 4))

# bins from 1.0 to 5.0 in 0.5-star increments
bins = np.arange(1.0, 5.5, 0.5)
plt.hist(yelp_df['stars'], bins=bins, edgecolor='black')

plt.xlabel('Stars (rating)')
plt.ylabel('Number of restaurants')
plt.title('Distribution of Yelp Restaurant Ratings (Stars)')
plt.xticks(np.arange(1.0, 5.5, 0.5))

plt.tight_layout()
plt.show()
```

Looks like our star ratings is left-skewed, which makes sense intuitively as ideally most restaurants would not average a 3.0 star rating! Instead our mean looks closer to 4.0.

Now let's take a look at the relationship between stars and review counts. Remember before when we log-transformed the `review_count` column? Let's take a look at why we might have done that.

```python
plt.subplot(1, 2, 1)
plt.scatter(df_rs['review_count'], df_rs['stars'], alpha=0.25, s=12)
plt.xlabel('Review count')
plt.ylabel('Stars')
plt.title('Stars vs Review Count (raw)')
```

As you can see, the outliers make it very difficult to properly visualize our data. Now let's use the log-transformed data:

```python
plt.subplot(1, 2, 2)
plt.scatter(df_rs['log_review_count'], df_rs['stars'], alpha=0.25, s=12)
plt.xlabel('log1p(Review count)')
plt.ylabel('Stars')
plt.title('Stars vs Review Count (log transform)')
```

We'd now like to answer the question, "Are higher star ratings associated with higher review counts?" To do that, we will take this data and run an OLS regression test, with `log_review_count` as our predictor variable and `stars` as our response variable.

```python
OLS_stars_reviews = smf.ols('stars ~ log_review_count', data=yelp_df).fit(cov_type='HC3')
print(OLS_stars_reviews.summary())
```

Looking at our OLS results, some things we can note:

- Our p-value is extremely small (much less than α<0.05), so there is a statistically significant chance that our results have an effect.
- Our effect size for log_review_count is 0.16, suggesting a small positive effect.
- Our R-squared value, at 0.052, suggests a small effect such that many other attributes are likely contributing to star number.

Overall, we've learned from this data that there is a statistically significant relationship between review counts and star ratings on Yelp's Open Dataset. The direct relationship is very small, however, so it is worth exploring other attributes.

### 1. Serving Alcohol and the Star Ratings
We'll start with alcohol; we coded our data earlier, now we can create plots to compare non-alcohol vs. alcohol providing restaurants.

```python
df_alc_known = yelp_df.dropna(subset=['alcohol_has'])
counts = df_alc_known['alcohol_has'].value_counts().sort_index()

plt.figure(figsize=(6, 4))
sns.boxplot(data=df_alc_known, x='alcohol_has', y='stars')
plt.xticks([0, 1], [f"No/NaN (n={counts.get(0,0)})", f"Has alcohol (n={counts.get(1,0)})"])
plt.xlabel("Alcohol")
plt.ylabel("Stars")
plt.title("Stars by Alcohol Availability (based on known entries)")
plt.tight_layout()
plt.show()
```

```python
model_alc = smf.ols('stars ~ alcohol_has', data=df_alc_known).fit(cov_type='HC3')
print(model_alc.summary())
```

Here are some notable points from the results:

- Our p-value is very large at 0.919, meaning the difference in star ratings from alcohol availability is very likely to have occurred by chance.
- Our R-squared value is extremely small, suggesting little-to-no effect.
  
We can clearly conclude from the data that alcohol is not a significant predictor for restaurant rating. There are likely other attributes which contribute more to a restaurant's rating than whether or not they serve alcohol!

### 3. WiFi and the Star Ratings
The digital world has nurtured food goers to appreciate WiFi services. 

```python
df_wifi = yelp_df.copy()

wifi_col = None
for c in ["attr_WiFi", "WiFi", "attributes.WiFi", "attributes_WiFi"]:
    if c in df_wifi.columns:
        wifi_col = c
        break

if wifi_col is None:
    raise KeyError("Couldn't find a WiFi column. Look for a column containing 'WiFi' in yelp_df.columns.")

wifi = df_wifi[wifi_col].astype("string")

wifi_clean = (
    wifi.str.lower()
        .str.strip()
        .str.replace(r"^u'", "", regex=True)     # u'free' -> free'
        .str.replace("'", "", regex=False)       # remove single quotes
        .str.replace('"', "", regex=False)       # remove double quotes if any
        .str.strip()
)

df_wifi["wifi_available"] = wifi_clean.map({
    "free": 1,
    "paid": 1,
    "no": 0,
    "none": 0
})

df_wifi_known = df_wifi.dropna(subset=["wifi_available"]).copy()

plt.figure(figsize=(6,4))
sns.boxplot(x="wifi_available", y="stars", data=df_wifi_known)
plt.xticks([0,1], ["No WiFi", "WiFi Available"])
plt.title("WiFi Availability vs Yelp Star Ratings")
plt.ylabel("Stars")
plt.xlabel("")
plt.show()
```
